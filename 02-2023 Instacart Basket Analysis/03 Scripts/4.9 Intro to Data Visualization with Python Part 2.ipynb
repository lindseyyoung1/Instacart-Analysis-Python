# 2. Looking Deeper into the Data to find connections

## 2.1 Histogram of the Hours of the day that orders are placed

```python
hist=orders_products_all['order_hour_of_day'].plot.hist(bins = 50)
plt.title('Order Hours of the Day', fontsize = 14, pad=20)
plt.ylabel('Frequency', fontsize = 12)
plt.xlabel('Hours',  fontsize = 12)
```

### Histogram shows that most orders are placed between hours 9 and 17, with the two highest hours being 10 and 11

## 2.2 Bar Chart of Order Frequency according to Day of the Week

```python
order_day_bar=orders_products_all['order_day_of_week'].value_counts().sort_index().plot.bar(color = ['lightsteelblue', 'cornflowerblue', 'dodgerblue', 'steelblue', 'blue', 'mediumblue', 'midnightblue'])
plt.title('Order Frequency according to Day of the Week', fontsize = 14, pad=20)
plt.ylabel('Frequency', fontsize = 12)
plt.xlabel('Days of the Week',  fontsize = 12)
```

## 2.3 Bar Chart showing the Customer Loyalty Distribution

```python
bar=orders_products_all['loyalty_flag'].value_counts().sort_index().plot.bar(color= ['yellowgreen','darkturquoise', 'salmon'])
plt.title('Loyalty Distribution', fontsize= 14)
plt.ylabel('Frequency (in millions)', fontsize = 12)
plt.xticks(rotation=0)
```

# 4. Scatterplot showing the Connection between age and income

```python
scatter_plot=sns.scatterplot(x = 'age', y = 'income',data = orders_products_all)
```

### According to the scatterplot, most individuals fall within an income level below 300000. Among the individuals who are above this amount, only a small portion are in their 20s, 30s, or 40s. There is a decent amount of people over the age of 40 who have an income of over 400000.

### This scatterplot shows that older individuals tend to have a higher income.
